```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
import spacy
```

```python
import pandas as pd
df = pd.read_csv('/content/Articles.csv', encoding='latin-1') # Try reading with 'latin-1' encoding
```

```python
df.head(5)
```

```python
nlp = spacy.load('en_core_web_sm')
def extract_dates_events(text):
    doc = nlp(text)
    dates = []
    events = []
    for ent in doc.ents:
        if ent.label_ == "DATE":
            dates.append(ent.text)
        elif ent.label_ == "EVENT":
            events.append(ent.text)
    return dates, events
df['Dates'], df['Events'] = zip(*df['Article'].apply(extract_dates_events))
output_file = '/content/Extracted_Dates_Events.csv'
df.to_csv(output_file, index=False)
extracted_df = pd.read_csv(output_file)
print("\nExtracted Dates and Events:")
print(extracted_df.head())
```

```python
print("\nExtracted Dates:")
for index, row in extracted_df.iterrows():
    print(f"Row {index + 1}: {row['Dates']}")

print("\nExtracted Events:")
for index, row in extracted_df.iterrows():
    print(f"Row {index + 1}: {row['Events']}")
```
